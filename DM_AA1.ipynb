```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('customer_data.csv')
```

```python
df.info()
```

```python
df.head(10)
```

```python
unique_customer_ids = df['Customer ID'].unique()
# Shuffle the unique customer IDs randomly
np.random.shuffle(unique_customer_ids)

# Create a mapping dictionary for membershiplevel and location
membershiplevel_mapping = {cust_id: np.random.choice(['Bronze', 'Silver', 'Gold', 'Platinum']) for cust_id in unique_customer_ids}
location_mapping = {cust_id: np.random.choice(['United States', 'Canada']) for cust_id in unique_customer_ids}

# Assign values to 'membershiplevel' and 'location' based on the mappings
df['MembershipLevel'] = df['Customer ID'].map(membershiplevel_mapping)
df['Location'] = df['Customer ID'].map(location_mapping)

df.head(20)
```

```python
# Set the percentage of values to change for each category
percentage_male_change = 20
percentage_female_change = 30
percentage_us_change = 30
percentage_ca_change = 20

# Change 'gender' column
male_indices_to_change = df[df['Gender'] == 'Male'].sample(frac=percentage_male_change/100).index
df.loc[male_indices_to_change, 'Gender'] = np.random.choice(['m', 'male', 'M'], size=len(male_indices_to_change))
female_indices_to_change = df[df['Gender'] == 'Female'].sample(frac=percentage_female_change/100).index
df.loc[female_indices_to_change, 'Gender'] = np.random.choice(['f', 'female', 'F'], size=len(female_indices_to_change))

# Change 'location' column
us_indices_to_change = df[df['Location'] == 'United States'].sample(frac=percentage_us_change/100).index
df.loc[us_indices_to_change, 'Location'] = 'US'

ca_indices_to_change = df[df['Location'] == 'Canada'].sample(frac=percentage_ca_change/100).index
df.loc[ca_indices_to_change, 'Location'] = 'CA'

df.head(10)
```
